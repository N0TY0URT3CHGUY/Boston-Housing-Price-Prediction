```python
#import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.datasets import load_boston
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor

#load the boston dataset
dataset = load_boston()

#Use dataframe to load the data
df = pd.DataFrame(dataset.data)

#Define X and y variable
X=dataset["data"]
y=dataset["target"]

#Split train and test data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=4)
X_train, X_validation, y_train, y_validation = train_test_split(X_test, y_test, test_size = 0.2, random_state = 4)

#Training and testing data
print("\n")
print("Original data\n\n",df.head(20))
print("After vectorized train data\n\n",X_train,"\n\n",y_train)
print("After vectorized test data\n\n",X_test)

#Gradient Boosting Regressor with tuned hyperparameters
reg = GradientBoostingRegressor(loss='ls', learning_rate=0.1, n_estimators=1000, subsample=1.0, 
                                criterion='friedman_mse', min_samples_split=2, min_samples_leaf=1,
                                min_weight_fraction_leaf=0.0, max_depth=3, min_impurity_decrease=0.0, 
                                min_impurity_split=None, init=None, random_state=None, max_features='auto',
                                alpha=0.9, verbose=0, max_leaf_nodes=None, warm_start=False, 
                                validation_fraction=0.1, n_iter_no_change=None, tol=0.0001, ccp_alpha=0.0)

#fitting the model
reg.fit(X_train, y_train)
reg.fit(X_test, y_test)
reg.fit(X_validation, y_validation)


# Predicting the Test, train, and validation results
y_pred = reg.predict(X_test)
y_pred1 = reg.predict(X_train)
y_pred2 = reg.predict(X_validation)


#Printing the accuracy values for the testing, training, and validation dataset
print("\nAccuracy of the testing dataset : ",reg.score(X_train , y_train)*100)
print("Accuracy of the training dataset: ",reg.score(X_test, y_test)*100)
print("Accuracy of the validation dataset: ",reg.score(X_validation, y_validation)*100)

#Top 10 incorrect predicted results
print("\nHere are the top 10 incorrect predicted results:")
print("\nThe predicted value for X[1] is:",list(reg.predict(X_test))[1])
print("The actual test value is:", list(y_test)[1])
print("The predicted value for X[3] is:", list(reg.predict(X_test))[3])
print("The actual test value is:", list(y_test)[3])
print("The predicted value for X[7] is:",list(reg.predict(X_test))[7])
print("The actual test value is:", list(y_test)[7])
print("The predicted value for X[8] is:", list(reg.predict(X_test))[8])
print("The actual test value is:", list(y_test)[8])
print("The predicted value for X[11] is:",list(reg.predict(X_test))[11])
print("The actual test value is:", list(y_test)[11])
print("The predicted value for X[17] is:", list(reg.predict(X_test))[17])
print("The actual test value is:", list(y_test)[17])
print("The predicted value for X[21] is:",list(reg.predict(X_test))[21])
print("The actual test value is:", list(y_test)[21])
print("The predicted value for X[23] is:", list(reg.predict(X_test))[23])
print("The actual test value is:", list(y_test)[23])
print("The predicted value for X[22] is:",list(reg.predict(X_test))[22])
print("The actual test value is:", list(y_test)[22])
print("The predicted value for X[27] is:", list(reg.predict(X_test))[27])
print("The actual test value is:", list(y_test)[27])
```
